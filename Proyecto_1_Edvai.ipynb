```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Utimas_Desvinculaciones_-_Row_data_1.csv', usecols = lambda column: column != 'Id')
```

Se procede a cambiar los tipos 'object' al tipo que corresponda. 

```python
data['Fecha comienzo'] = pd.to_datetime(data['Fecha comienzo'])
data['Fecha fin'] = pd.to_datetime(data['Fecha fin'])
```

```python
def extract_number_level(s):
    return int(s[1:]) 
```

```python
data['Nivel'] = data['Nivel'].apply(extract_number_level)
```

```python
def extract_number_salary(s):
    return int(s[5:]) 
```

```python
data['Rango Salarial'] = data['Rango Salarial'].apply(extract_number_salary)
data['RS competencia'] = data['RS competencia'].apply(extract_number_salary)
```

```python
data['Realizo Cursos'] = data['Realizo Cursos'].replace({'SI': 1, 'NO': 0})
```

```python
data
```

### Datos agregados
Se utilizarán métodos correspondientes a la libreria pandas para calcular datos de interés. En primera instancia se agregará la cantidad de tiempo laboral, restando la fecha de salida menos la de entrada y el tiempo recruitment. 

```python
data['Duracion con recruitment'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days/29.5
```

```python
data['Duracion sin recruitment'] = (data['Fecha fin'] - data['Fecha comienzo']).dt.days/29.5 - data['Tiempo recruitment']
```

```python
grupos_unicos = data['Grupo recruitment'].unique()
grupo_numero_map = {grupo: i for i, grupo in enumerate(grupos_unicos)}
grupo_numero_map
```

```python
data['Grupo_Numero'] = data['Grupo recruitment'].map(grupo_numero_map)
data['Grupo_Numero'] 
```

Se agrega un mapeo de las areas a números para poder relacionar la variable con las demás en la matriz de dispersión. 

```python
# Crear un diccionario que mapee cada área única a un número único
areas_unicas = data['Area'].unique()
area_numero_map = {area: i for i, area in enumerate(areas_unicas)}
area_numero_map
```

```python
# Mapear los nombres de áreas a números en una nueva columna llamada 'Area_Numero'
data['Area_Numero'] = data['Area'].map(area_numero_map)
data['Area_Numero'] 
```
